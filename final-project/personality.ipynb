Create a function to replace the missing values: 
- by 0 if we expected a number between 1 to 5 as Rock, Alternative, Borrowed stuff, Biologie ...
- by the average if we expected a "real" number as age, weight ...
- by "no value" is we expected a string
- by "no" if we expected a boolean value, we do the hypothese that no answer correspond to a "no"


```python
import math
```

```python
def replace_missing_value(row):
    for column in personalities.columns:
        if personalities[column].dtypes == object:
            if type(row[column]) == float:
                personalities[column] = personalities[column].fillna(value='no value')
        elif personalities[column].dtypes == float or personalities[column].dtypes == int:
            if math.isnan(row[column]):
                personalities[column] = personalities[column].fillna(value=personalities[column].mean())
            if row[column] == 0:
                personalities[column] = personalities[column].fillna(value=0)
                  
```
