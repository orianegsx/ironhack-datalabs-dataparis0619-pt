#### List all language names and number of related articles in the order they appear in wikipedia.org

````python
# This is the url you will scrape in this exercise
url = 'https://www.wikipedia.org/'
````

````python
#your code

html = requests.get(url).content
soup = BeautifulSoup(html, "lxml")
div_tags = soup.find_all('div', {'data-el-section':'secondary links'})


languages = {}
# la fonction .get permet de trouver un attribut dans un tag (librairie Beautiful Soup)
for i, divtag in enumerate(div_tags):
    a_tags = divtag.find_all('a')
    languages1=[]
    for a_tag in a_tags:
        #print(a_tag.text)
        #print(i)
        languages1.append(a_tag.text)
        #language = a_tag.text
        #languages1[i] = []
        #languages1[i].append(language)
        languages[i]=languages1

languages
````
