1. Import librairies

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
from pprint import pprint
from lxml import html
from lxml.html import fromstring
import urllib.request
from urllib.request import urlopen
import random
import re
import scrapy
```

2. Request url and get the content

```python
url = 'https://fr.wikipedia.org/wiki/Game_of_Thrones'
```

```python
html = requests.get(url).content
soup = BeautifulSoup(html, "lxml")
soup
```

3. Save the content into a file

```python
with open("output.html", "w") as file:
    file.write(str(soup))
```

4. Get all the actors' name, the url link, vf and their characters' name 

```python
table_tags = soup.find_all('table') #On trouve toutes les tables, on voit que celle qu'on cherche c'est la 2eme ("[1]")
trs = table_tags[1].find_all("tr") #On trouve toutes les lignes tr

td_list = [x.find_all('td') for x in trs if x.find_all('td')] #Chaque tr a toujours 3 premiers td
#td_list[0][2].text

```

```python
def clean(x):
    var = x.replace('«\xa0', '')
    var = var.replace('\xa0»', '')
    var = var.replace('<.*?>', '')
    return var.strip()


actors = [(clean(str(x[0].text)), clean(str(x[0].find('a').get('href'))), clean(str(x[1].text)), clean(str(x[2].text))) for x in td_list]


actors
```

5. Transform your data into a dataFrame

```python
df = pd.DataFrame(actors)
df
```

6. Add for each column the description of the actor on his personal page

```python
df[1] = df[1].map(lambda x : 'https://fr.wikipedia.org' + x)
```

```python
paragraphes = []
for url in df[1]:
    response1 = requests.get(url).content
    soup1 = BeautifulSoup(response1, "lxml")
    para = soup1.find('time',{'class':'nowrap date-lien bday'})
    paragraphes.append(para)

df[4] = paragraphes
df
```

8. Create a function to clean html tag from your paragraph

```python
def remove_html_tags(text):
    """Remove html tags from a string"""
    import re
    clean = re.compile('<.*?>')
    return re.sub(clean, '', text)

#Exemple
string = '<time class="nowrap date-lien bday" datetime="1964-01-14" data-sort-value="1964-01-14"><a href="/wiki/14_janvier" title="14 janvier">14</a> <a href="/wiki/Janvier_1964" title="Janvier 1964">janvier</a> <a href="/wiki/1964" title="1964">1964</a></time>'

remove_html_tags(string)
```

```python
df['birthday'] = df['birthday'].map(lambda x : remove_html_tags(str(x)))
```
