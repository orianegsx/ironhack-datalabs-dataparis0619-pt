```python
import pandas as pd
import numpy as np
import os
from scipy import stats
from scipy.stats import chi2_contingency
```

```python
astro_metier=pd.read_csv('astro_metier.csv')
astro_metier
```

```python

contigency = pd.crosstab(astro_metier['element'], astro_metier['categorie'])
contigency
```

```python
chi2, p, dof, ex = chi2_contingency(contigency, correction=False)
```
