#### Using Seaborn plot the correlation matrix of various features.

*Hint: search how to use `heatmap`*

```python
??sns.heatmap
```

```python
#your code here
corr = titanic.corr()
ax = sns.heatmap(
    corr, 
    annot=True,
    vmin=-1, vmax=1, center=0,
    cmap=sns.diverging_palette(20, 220, n=200),
    square=True
)
ax.set_xticklabels(
    ax.get_xticklabels(),
    rotation=45,
    horizontalalignment='right'
);
```
