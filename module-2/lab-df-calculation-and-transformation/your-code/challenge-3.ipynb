```python
def get_set(serie_):
    return set(serie_)

agg_funcs = {'amount_spent':sum, 'InvoiceNo':get_set}

agg_funcs = {'amount_spent':sum}

df = orders.groupby('CustomerID').agg(agg_funcs)

```

```python
quant_75 = df["amount_spent"].quantile(0.75)
```

```python
quant_95 = df["amount_spent"].quantile(0.95)
quant_95
```

```python
#df["label"] = df[df["amount_spent"] > quant_95, "VIP", "classique"]

df["label"] = np.where(df["amount_spent"] > quant_95, "VIP", np.where(df["amount_spent"] > quant_75, "Preferred", "Classique"))
df.head()


```
